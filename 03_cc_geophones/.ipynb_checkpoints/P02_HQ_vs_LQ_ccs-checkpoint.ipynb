## Plot correlation coefficients between waveforms for low quality vs. high quality detections

### Import libraries

```python
import pandas as pd
from datetime import datetime, timedelta
import numpy as np
import os
import glob
import sys
import matplotlib.pyplot as plt

# Import functions
fxndir = '../functions/'
sys.path.insert(0,fxndir)
from moon2data import *
```

### Load catalog of detections

```python
# Load combined catalog
mqdir1 = '../catalogs/quality_control/'
cat = pd.read_csv(mqdir1 + 'A17_moonquakes_catalog_nodupes.csv')
cat.drop(list(cat.filter(regex='Unnamed|index')), axis=1, inplace=True)

# High quality vs low quality
cat_hq = cat.loc[cat.quality == 'HQ'].reset_index()
inds_hq = np.arange(0,len(cat_hq)).tolist()
cat_lq = cat.loc[cat.quality == 'LQ'].reset_index()
inds_lq = np.arange(0,len(cat_lq)).tolist()
```

### Load catalog of cross-correlation between geophones

```python
mqdir2 = '../catalogs/cc_geophones/'
cc_cat = pd.read_csv(mqdir2 + 'A17_cc_geophones_catalog.csv')
```

### Add information on event quality

```python
# Iterate through events
event_grades = []
for r in np.arange(0,len(cc_cat)):
    row = cc_cat.iloc[r]
    evid = row.evid
    row_quality = cat.loc[cat.evid == evid].iloc[0]
    event_grades.append(row_quality.grade_new)
    
# Add information on event quality and save
cc_cat['grade_new'] = event_grades
cc_cat.to_csv(mqdir2 + 'A17_cc_geophones_catalog.csv',index=False)
```

### Determine quality of geophone pairs

```python
pairs = ['12', '13', '14', '23', '34', '24']
evids = np.unique(cat.evid.tolist())
evids_list = []
ccs_list = []
dts_list = []
pairs_list = []
qualities_list = []
grades_list = []

# Iterate through events
for evid in evids:
    cc_row = cc_cat.loc[cc_cat.evid == evid].iloc[0]
    for pair in pairs:
        cc = cc_row[f'cc_{pair}']
        dt = cc_row[f'dt_{pair}']
        geoA = int(pair[0])
        geoB = int(pair[1])
        rows = cat.loc[(cat.evid == evid) & ((cat.geophone == geoA) | (cat.geophone == geoB))]
        num_lq = len(rows.loc[rows.quality == 'LQ'])
        if num_lq == 2:
            quality = 'LQ-LQ'
        elif num_lq == 1:
            quality = 'HQ-LQ'
        elif num_lq == 0:
            quality = 'HQ-HQ'
        grade = cc_row.grade_new

        # Append
        evids_list.append(evid)
        pairs_list.append(pair)
        ccs_list.append(cc)
        dts_list.append(dt)
        qualities_list.append(quality)
        grades_list.append(grade)
```

```python
# Save in dataframe
d = {'evid':evids_list, 'pair':pairs_list, 'cc':ccs_list, 'dt':dts_list, 
    'quality':qualities_list, 'grade_new':grades_list}
quality_df = pd.DataFrame(data = d)
quality_df.to_csv(mqdir2 + 'A17_cc_geophones_catalog_quality.csv',index=False)
```

### Plot lag and correlation coefficient between geophones

```python
# Break up into pairs
# 1-2
df_12 = quality_df.loc[quality_df.pair == '12']
lq_df_12 = df_12.loc[df_12.quality == 'LQ-LQ']
hq_df_12 = df_12.loc[df_12.quality == 'HQ-HQ']
lq_lags_12 = np.array(lq_df_12.dt.tolist())
lq_ccs_12 = np.array(lq_df_12.cc.tolist())
hq_lags_12 = np.array(hq_df_12.dt.tolist())
hq_ccs_12 = np.array(hq_df_12.cc.tolist())

# 1-3
df_13 = quality_df.loc[quality_df.pair == '13']
lq_df_13 = df_13.loc[df_13.quality == 'LQ-LQ']
hq_df_13 = df_13.loc[df_13.quality == 'HQ-HQ']
lq_lags_13 = np.array(lq_df_13.dt.tolist())
lq_ccs_13 = np.array(lq_df_13.cc.tolist())
hq_lags_13 = np.array(hq_df_13.dt.tolist())
hq_ccs_13 = np.array(hq_df_13.cc.tolist())

# 1-4
df_14 = quality_df.loc[quality_df.pair == '14']
lq_df_14 = df_14.loc[df_14.quality == 'LQ-LQ']
hq_df_14 = df_14.loc[df_14.quality == 'HQ-HQ']
lq_lags_14 = np.array(lq_df_14.dt.tolist())
lq_ccs_14 = np.array(lq_df_14.cc.tolist())
hq_lags_14 = np.array(hq_df_14.dt.tolist())
hq_ccs_14 = np.array(hq_df_14.cc.tolist())

# 2-3
df_23 = quality_df.loc[quality_df.pair == '23']
lq_df_23 = df_23.loc[df_23.quality == 'LQ-LQ']
hq_df_23 = df_23.loc[df_23.quality == 'HQ-HQ']
lq_lags_23 = np.array(lq_df_23.dt.tolist())
lq_ccs_23 = np.array(lq_df_23.cc.tolist())
hq_lags_23 = np.array(hq_df_23.dt.tolist())
hq_ccs_23 = np.array(hq_df_23.cc.tolist())

# 3-4
df_34 = quality_df.loc[quality_df.pair == '34']
lq_df_34 = df_34.loc[df_34.quality == 'LQ-LQ']
hq_df_34 = df_34.loc[df_34.quality == 'HQ-HQ']
lq_lags_34 = np.array(lq_df_34.dt.tolist())
lq_ccs_34 = np.array(lq_df_34.cc.tolist())
hq_lags_34 = np.array(hq_df_34.dt.tolist())
hq_ccs_34 = np.array(hq_df_34.cc.tolist())

# 2-4
df_24 = quality_df.loc[quality_df.pair == '24']
lq_df_24 = df_24.loc[df_24.quality == 'LQ-LQ']
hq_df_24 = df_24.loc[df_24.quality == 'HQ-HQ']
lq_lags_24 = np.array(lq_df_24.dt.tolist())
lq_ccs_24 = np.array(lq_df_24.cc.tolist())
hq_lags_24 = np.array(hq_df_24.dt.tolist())
hq_ccs_24 = np.array(hq_df_24.cc.tolist())
```

```python
# Initialize figure
fig,axs = plt.subplots(2,3,figsize=(12,5),dpi=500)

# 1-2
ax12 = axs[0][0]
n,bins,_ = ax12.hist(lq_lags_12,bins=40,alpha=0.5,edgecolor='black',label='LQ-LQ')
ax12.hist(hq_lags_12,bins=bins,alpha=0.5,edgecolor='black',label='HQ-HQ')
ax12.set_xlim([-4,4])
ax12.set_ylabel('# Pairs',fontsize=11)
ax12.set_xlabel('Lag (s)',fontsize=11)
ax12.tick_params(axis='both',labelsize=11)
ax12.set_title('GEO1 - GEO2',fontsize=12,fontweight='bold')
ax12.legend(loc='upper left')

# 1-3
ax13 = axs[0][1]
ax13.hist(lq_lags_13,bins=bins,alpha=0.5,edgecolor='black')
ax13.hist(hq_lags_13,bins=bins,alpha=0.5,edgecolor='black')
ax13.set_xlim([-4,4])
ax13.set_ylabel('# Pairs',fontsize=11)
ax13.set_xlabel('Lag (s)',fontsize=11)
ax13.tick_params(axis='both',labelsize=11)
ax13.set_title('GEO1 - GEO3',fontsize=12,fontweight='bold')

# 1-4
ax14 = axs[0][2]
ax14.hist(lq_lags_14,bins=bins,alpha=0.5,edgecolor='black')
ax14.hist(hq_lags_14,bins=bins,alpha=0.5,edgecolor='black')
ax14.set_xlim([-4,4])
ax14.set_ylabel('# Pairs',fontsize=11)
ax14.set_xlabel('Lag (s)',fontsize=11)
ax14.tick_params(axis='both',labelsize=11)
ax14.set_title('GEO1 - GEO4',fontsize=12,fontweight='bold')

# 2-3
ax23 = axs[1][0]
ax23.hist(lq_lags_23,bins=bins,alpha=0.5,edgecolor='black')
ax23.hist(hq_lags_23,bins=bins,alpha=0.5,edgecolor='black')
ax23.set_xlim([-4,4])
ax23.set_ylabel('# Pairs',fontsize=11)
ax23.set_xlabel('Lag (s)',fontsize=11)
ax23.tick_params(axis='both',labelsize=11)
ax23.set_title('GEO2 - GEO3',fontsize=12,fontweight='bold')

# 3-4
ax34 = axs[1][1]
ax34.hist(lq_lags_34,bins=bins,alpha=0.5,edgecolor='black')
ax34.hist(hq_lags_34,bins=bins,alpha=0.5,edgecolor='black')
ax34.set_xlim([-4,4])
ax34.set_ylabel('# Pairs',fontsize=11)
ax34.set_xlabel('Lag (s)',fontsize=11)
ax34.tick_params(axis='both',labelsize=11)
ax34.set_title('GEO3 - GEO4',fontsize=12,fontweight='bold')

# 2-4
ax24 = axs[1][2]
ax24.hist(lq_lags_24,bins=bins,alpha=0.5,edgecolor='black')
ax24.hist(hq_lags_24,bins=bins,alpha=0.5,edgecolor='black')
ax24.set_xlim([-4,4])
ax24.set_ylabel('# Pairs',fontsize=11)
ax24.set_xlabel('Lag (s)',fontsize=11)
ax24.tick_params(axis='both',labelsize=11)
ax24.set_title('GEO2 - GEO4',fontsize=12,fontweight='bold')

# Subplots
plt.subplots_adjust(hspace = 0.5,wspace = 0.35)
plt.show()
```
