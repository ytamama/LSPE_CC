## Plot a map of LSPE instruments

### Import libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
import math
```

### Load coordinates of geophones and LM

```python
mqdir = '../catalogs/coordinates/'
loctbl = pd.read_csv(mqdir + 'Apollo17_Cartesian_Coordinates.csv')

# Unpack coordinates
xvals = np.array(loctbl.x_m.tolist()[0:-1])
yvals = np.array(loctbl.y_m.tolist()[0:-1])
```

### Load coordinates of boulders

```python
# Boulders
boulder_cat = pd.read_csv(mqdir + 'Apollo17_boulder_coordinates.csv')
boulder_xs = np.array(boulder_cat.x_m.tolist())
boulder_ys = np.array(boulder_cat.y_m.tolist())
boulder_diams = np.array(boulder_cat.diameter_m.tolist())
```

### Plot coordinates of geophones and boulders

```python
fig,ax = plt.subplots(1,1,figsize=(5, 5.5), dpi=500)
ax.set_xlabel('X (m)',fontsize=12)
ax.set_ylabel('Y (m)',fontsize=12)
for b in np.arange(0,len(boulder_diams)):
    if (b == 0):
        ax.scatter(boulder_xs[b],boulder_ys[b],s=boulder_diams[b]*boulder_diams[b]*3.5,color='black',label='6 m')
    elif (b == 2):
        ax.scatter(boulder_xs[b],boulder_ys[b],s=boulder_diams[b]*boulder_diams[b]*3.5,color='black',label='3 m')
    else:
        ax.scatter(boulder_xs[b],boulder_ys[b],s=boulder_diams[b]*boulder_diams[b]*3.5,color='black')
ax.scatter(xvals,yvals,s=5)
ax.legend(loc='lower right',fontsize=9)
ax.tick_params(axis='both',labelsize=12)
ax.set_aspect('equal')
```
