### 1 - Plot demonstrating correlation coefficients

#### Load example catalog of cross-correlation results

```python
catdir = '../catalogs/cc_moonquakes_sample/'
cc_cat = pd.read_csv(catdir + 'AlignArrivals_CC_REF-761226-15-M1.csv')
```

#### Select events with a correlation coefficient of 1, 0.9, and 0.5

```python
# CC = 1
cc_cat1 = cc_cat.loc[cc_cat.corr_coeffs >= 0.99]
evtrow1 = cc_cat1.iloc[0]

# CC = 0.90
cc_cat2 = cc_cat.loc[(cc_cat.corr_coeffs >= 0.9) & (cc_cat.corr_coeffs <= 0.9005)]
evtrow2 = cc_cat2.iloc[0]

# CC = 0.50
cc_cat3 = cc_cat.loc[(cc_cat.corr_coeffs >= 0.49) & (cc_cat.corr_coeffs <= 0.51)]
evtrow3 = cc_cat3.iloc[0]
```

#### Obtain waveforms for each event

```python
# Inputs
befwin = 5
aftwin = 15
parentdir = '/data/ytamama/Apollo17/LSPE_data/sac_volts_ds/'
minfreq = 5
maxfreq = 35

# CC = 1.0
arrtime1 = datetime.strptime(evtrow1.mod_arrival_time, '%Y-%m-%d %H:%M:%S.%f')
st1 = moon2sac(arrtime1,1,befwin,aftwin,minfreq,maxfreq,parentdir)
trtimes1 = st1.traces[0].times()
trdata1 = st1.traces[0].data
trdata1_norm = trdata1 / np.max(np.abs(trdata1))

# CC = 0.9
arrtime2 = datetime.strptime(evtrow2.mod_arrival_time, '%Y-%m-%d %H:%M:%S.%f')
st2 = moon2sac(arrtime2,1,befwin,aftwin,minfreq,maxfreq,parentdir)
trtimes2 = st2.traces[0].times()
trdata2 = st2.traces[0].data
trdata2_norm = trdata2 / np.max(np.abs(trdata2))

# CC = 0.51
arrtime3 = datetime.strptime(evtrow3.mod_arrival_time, '%Y-%m-%d %H:%M:%S.%f')
st3 = moon2sac(arrtime3,1,befwin,aftwin,minfreq,maxfreq,parentdir)
trtimes3 = st3.traces[0].times()
trdata3 = st3.traces[0].data
trdata3_norm = trdata3 / np.max(np.abs(trdata3))
```

#### Plot waveforms

```python
# 
fig,axs = plt.subplots(1,3,figsize=(12,3.),dpi=500)

# CC = 1
ax1 = axs[0]
ax1.plot(trtimes1,trdata1_norm+2,color='red')
ax1.plot(trtimes1,trdata1_norm,color='black')
ax1.set_xlim([0,befwin+aftwin])
ax1.set_yticks([])
ax1.set_xlabel('Time (s)',fontsize=12)
ax1.set_ylabel('Normalized volts',fontsize=12)
ax1.tick_params(axis='both',labelsize=12)
ax1.set_title('CC: 1.0',fontsize=13,fontweight='bold')

# CC = 0.65
ax2 = axs[1]
ax2.plot(trtimes1,trdata1_norm+2,color='red')
ax2.plot(trtimes2,trdata2_norm,color='black')
ax2.set_xlim([0,befwin+aftwin])
ax2.set_yticks([])
ax2.set_xlabel('Time (s)',fontsize=12)
ax2.set_ylabel('Normalized volts',fontsize=12)
ax2.tick_params(axis='both',labelsize=12)
ax2.set_title('CC: 0.90',fontsize=13,fontweight='bold')

# CC = 0.35
ax3 = axs[2]
ax3.plot(trtimes1,trdata1_norm+2,color='red')
ax3.plot(trtimes3,trdata3_norm,color='black')
ax3.set_xlim([0,befwin+aftwin])
ax3.set_yticks([])
ax3.set_xlabel('Time (s)',fontsize=12)
ax3.set_ylabel('Normalized volts',fontsize=12)
ax3.tick_params(axis='both',labelsize=12)
ax3.set_title('CC: 0.51',fontsize=13,fontweight='bold')
```

### Demos of SNR function and duration
#### Load catalog of repeating moonquakes

```python
# Catalog with information by geophone
mqdir1 = '../catalogs/final_catalogs/isolated_vs_repeating/'
repeating_cat1 = pd.read_csv(mqdir1 + 'A17_repeating_catalog_HQ_final.csv')
```

#### Inputs to obtain waveforms

```python
parentdir = '/data/ytamama/Apollo17/LSPE_data/sac_volts_ds/'
minfreq = 3
maxfreq = 35
befwin1 = 60
aftwin1 = 60
befwin2 = 60
aftwin2 = 600
geonum = 1
```

#### Select an event and obtain characteristics

```python
# Arrival time, emergence, duration
evid = '761226-15-M1'
row = repeating_cat1.loc[(repeating_cat1.evid == evid) & (repeating_cat1.geophone == 1)].iloc[0]
arrtime = datetime.strptime(row.picktime_SNR,'%Y-%m-%d %H:%M:%S.%f')
emergence = row.emergence_s
duration = row.duration_s

# RMSE
mqdir2 = '../catalogs/quality_control/'
cat2 = pd.read_csv(mqdir2 + 'A17_moonquakes_catalog_nodupes.csv')
row2 = cat2.loc[(cat2.evid == evid) & (cat2.geophone == 1)].iloc[0]
rmse = row2.rmse_gauss
```

#### Obtain waveform and envelope

```python
# Short
st1 = moon2sac(arrtime,geonum,befwin1,aftwin1,minfreq,maxfreq,parentdir)
trtimes1 = st1.traces[0].times() - befwin1
trdata1 = st1.traces[0].data
trdata_env1 = obspy.signal.filter.envelope(trdata1)

# Long
st2 = moon2sac(arrtime,geonum,befwin2,aftwin2,minfreq,maxfreq,parentdir)
trtimes2 = st2.traces[0].times() - befwin2
trdata2 = st2.traces[0].data
trdata_env2 = obspy.signal.filter.envelope(trdata2)
```

#### Calculate SNR function

```python
winlen = 10
stepby = 0.1

# Short
snrtimes1, snrvals1 = snr_window(trdata_env1,trtimes1,winlen,stepby)
# Fit Gaussian to SNR function
rmse1, snrvals_gauss_norm1, snrvals_norm1, params1 = fit_gaussian(snrvals1,snrtimes1)
```

#### Plot: SNR function

```python
# Initialize figure
fig = plt.figure(figsize=(12, 3.5), dpi=500)
gs = gridspec.GridSpec(2, 3) 

# Short waveform
ax1 = fig.add_subplot(gs[0,0])
ax1.plot(trtimes1,trdata1,color='black')
ax1.set_xlim([-30,40])
ax1.axvline(x = 0, color='red', linestyle='--',linewidth=2)
ax1.axvline(x = emergence, color='red', linestyle='--',linewidth=2)
ax1.set_ylabel('Volts',fontsize=11)
ax1.tick_params(axis='both',labelsize=10.5)

# Short envelope
ax2 = fig.add_subplot(gs[0,1])
ax2.plot(trtimes1,trdata_env1,color='black')
ax2.set_xlim([-30,40])
ax2.axvline(x = 0, color='red', linestyle='--',linewidth=2)
ax2.axvline(x = emergence, color='red', linestyle='--',linewidth=2)
ax2.set_ylabel('Volts',fontsize=11)
ax2.set_title(f'Event {evid}',fontsize=12,fontweight='bold')
ax2.tick_params(axis='both',labelsize=10.5)

# SNR function
ax3 = fig.add_subplot(gs[0,2])
ax3.plot(snrtimes1,snrvals_norm1,color='black',label='SNR fxn',linewidth=2)
ax3.plot(snrtimes1,snrvals_gauss_norm1,color='red',label=f'RMSE: {np.round(rmse,3)}',linewidth=2)
ax3.axvline(x = 0, color='grey', linestyle='--')
ax3.set_xlim([-30,40])
ax3.legend(loc='upper right',fontsize=9)
ax3.set_ylabel('SNR (Norm.)',fontsize=11)
ax3.tick_params(axis='both',labelsize=10.5)

# SNR of whole waveform
ax4 = fig.add_subplot(gs[1,:])
ax4.plot(trtimes1,trdata_env1,color='black')
ax4.set_xlim([-60,60])
ax4.axvline(x = 0, linestyle='--', color='grey')
ax4.set_ylabel('Volts',fontsize=11)
ax4.tick_params(axis='both',labelsize=10.5)
ax4.set_xlabel('Time (s)',fontsize=11)
# 
plt.subplots_adjust(wspace = 0.35,hspace = 0.3)
```

#### Plot: duration

```python
# Initialize figure
fig = plt.figure(figsize=(12, 3.5),dpi=500)
gs = gridspec.GridSpec(2, 3) 

# Plot duration
ax1 = fig.add_subplot(gs[0,:])
ax1.plot(trtimes2,trdata2,color='black')
ax1.axvline(x = 0,color='red',linewidth=1.5,linestyle='--')
ax1.axvline(x = duration,color='red',linewidth=1.5,linestyle='--')
ax1.set_xlim([-10,190])
ax1.set_ylabel('Volts',fontsize=11)
ax1.tick_params(axis='both',labelsize=10.5)

# Duration -- envelope
ax2 = fig.add_subplot(gs[1,:])
ax2.plot(trtimes2,trdata_env2,color='black')
ax2.axvline(x = 0,color='red',linewidth=1.5,linestyle='--')
ax2.axvline(x = duration,color='red',linewidth=1.5,linestyle='--')
ax2.set_xlim([-10,190])
ax2.set_xlabel('Time (s)',fontsize=11)
ax2.set_ylabel('Volts',fontsize=11)
ax2.tick_params(axis='both',labelsize=10.5)
plt.subplots_adjust(wspace = 0.35,hspace = 0.3)
```

### SNR function demo for a "low-quality" moonquake

```python
# Load catalog
cat = pd.read_csv(mqdir2 + 'A17_moonquakes_catalog_nodupes.csv')
evid = '770202-06-N1'
row_lq = cat.loc[(cat.evid == evid) & (cat.geophone == 3)].iloc[0]
arrtime_lq = datetime.strptime(row_lq.picktime,'%Y-%m-%d %H:%M:%S.%f')

# Obtain waveform + envelope
geonum = 3
st_lq = moon2sac(arrtime_lq,geonum,befwin1,aftwin1,minfreq,maxfreq,parentdir)
trtimes_lq = st_lq.traces[0].times() - befwin1
trdata_lq = st_lq.traces[0].data
trdata_env_lq = obspy.signal.filter.envelope(trdata_lq)
trdata_env_lq = trdata_env_lq - np.min(trdata_env_lq)
snrtimes_lq, snrvals_lq = snr_window(trdata_env_lq,trtimes_lq,winlen,stepby)
rmse_lq, snrvals_gauss_norm_lq, snrvals_norm_lq, params_lq = fit_gaussian(snrvals_lq,snrtimes_lq)

### Plot
# Initialize figure
fig = plt.figure(figsize=(12, 6),dpi=500)
gs = gridspec.GridSpec(3, 1) 

# Short waveform
ax1 = fig.add_subplot(gs[0,0])
ax1.plot(trtimes_lq,trdata_lq,color='black')
ax1.set_xlim([-40,40])
ax1.axvline(x = 0, color='red', linestyle='--',linewidth=2)
ax1.set_ylabel('Volts',fontsize=11)
ax1.tick_params(axis='both',labelsize=10.5)
ax1.set_title(f'Event {evid}',fontsize=12,fontweight='bold')

# Short envelope
ax2 = fig.add_subplot(gs[1,0])
ax2.plot(trtimes_lq,trdata_env_lq,color='black')
ax2.set_xlim([-40,40])
ax2.axvline(x = 0, color='red', linestyle='--',linewidth=2)
ax2.set_ylabel('Volts',fontsize=11)
ax2.tick_params(axis='both',labelsize=10.5)

# SNR function
ax3 = fig.add_subplot(gs[2,0])
ax3.plot(snrtimes_lq,snrvals_norm_lq,color='black',label='SNR fxn',linewidth=2)
ax3.plot(snrtimes_lq,snrvals_gauss_norm_lq,color='red',label=f'RMSE: {np.round(rmse_lq,3)}',linewidth=2)
ax3.axvline(x = 0, color='grey', linestyle='--')
ax3.set_xlim([-40,40])
ax3.legend(loc='upper right',fontsize=9,framealpha=1,ncol=2)
ax3.set_xlabel('Time (s)',fontsize=11)
ax3.set_ylabel('SNR (Norm.)',fontsize=11)
ax3.tick_params(axis='both',labelsize=10.5)
# 
plt.subplots_adjust(wspace = 0.35,hspace = 0.26)
```
