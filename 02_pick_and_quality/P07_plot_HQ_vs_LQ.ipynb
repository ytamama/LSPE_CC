## Plot examples of high quality vs. low quality waveforms

### Import libraries

```python
import pandas as pd
from obspy import read,UTCDateTime
from datetime import datetime, timedelta
import numpy as np
import os
import glob
import sys
import matplotlib.pyplot as plt
import random

# Import functions
fxndir = '../functions/'
sys.path.insert(0,fxndir)
from moon2data import *
```

### Load moonquake catalog 

```python
# Load catalog of moonquakes
mqdir = '../catalogs/quality_control/'
cat = pd.read_csv(mqdir + 'A17_moonquakes_catalog_nodupes.csv')
cat.drop(list(cat.filter(regex='Unnamed|index')), axis=1, inplace=True)
```

### Inputs to obtain waveforms

```python
parentdir = '/data/ytamama/Apollo17/LSPE_data/sac_volts_ds/'
minfreq = 3
maxfreq = 35
befwin = 60
aftwin = 60
```

### Select 4 high quality and 4 low quality

```python
# High quality -- mix of emergent and impulsive events
evids_hq = ['761130-01-M4','770324-09-M1','761207-21-M4','770125-07-M1']
geonums_hq = [2, 1, 2, 3]

# Low-quality -- events characterized by spikes or obscured by noise
evids_lq = ['770127-05-N1','770301-01-N2','770207-20-M7','770202-06-N1']
geonums_lq = [4, 2, 2, 3]
```

### Plot waveforms

```python
# Initialize figure
fig,axs = plt.subplots(4,2,figsize=(12,9),dpi=500)

# High quality
for i in np.arange(0,len(evids_hq)):
    evid = evids_hq[i]
    geonum = geonums_hq[i]
    row = cat.loc[(cat.evid == evid) & (cat.geophone == geonum)].iloc[0]
    arrtime = datetime.strptime(row.picktime,'%Y-%m-%d %H:%M:%S.%f')
    geonum = row.geophone
    evid = row.evid
    rmse = np.round(row.rmse_gauss,3)
    snr = np.round(row.SNR,3)
    gap_len = np.round(row.max_gap_len_s,3)
    quotient = np.round(row.SNR_div_RMSE,3)
    
    # Obtain seismogram
    st = moon2sac(arrtime,geonum,befwin,aftwin,minfreq,maxfreq,parentdir)
    trtimes = st.traces[0].times()
    trdata = st.traces[0].data
    
    # Plot waveform
    ax = axs[i][0]
    ax.plot(trtimes,trdata,color='black',label='GEO' + str(geonum))
    ax.set_xlim([0,befwin+aftwin])
    ax.set_ylabel('Volts',fontsize=11)
    ax.tick_params(axis='both',labelsize=10)
    ax.legend(loc='upper left')
    ax.set_title(f'{evid} \nRMSE: {rmse}, SNR: {snr}, Gap: {gap_len}s, SNR/RMSE: {quotient}',fontsize=11)
    if i == len(evids_hq)-1:
        ax.set_xlabel('Time (s)',fontsize=11)
    
    
# Low quality
for i in np.arange(0,len(evids_lq)):
    evid = evids_lq[i]
    geonum = geonums_lq[i]
    row = cat.loc[(cat.evid == evid) & (cat.geophone == geonum)].iloc[0]
    arrtime = datetime.strptime(row.picktime,'%Y-%m-%d %H:%M:%S.%f')
    geonum = row.geophone
    evid = row.evid
    rmse = np.round(row.rmse_gauss,3)
    snr = np.round(row.SNR,3)
    gap_len = np.round(row.max_gap_len_s,3)
    quotient = np.round(row.SNR_div_RMSE,3)
    
    # Obtain seismogram
    st = moon2sac(arrtime,geonum,befwin,aftwin,minfreq,maxfreq,parentdir)
    trtimes = st.traces[0].times()
    trdata = st.traces[0].data
    
    # Plot waveform
    ax = axs[i][1]
    ax.plot(trtimes,trdata,color='black',label='GEO' + str(geonum))
    ax.set_xlim([0,befwin+aftwin])
    ax.set_ylabel('Volts',fontsize=11)
    ax.tick_params(axis='both',labelsize=10)
    ax.legend(loc='upper left')
    ax.set_title(f'{evid} \nRMSE: {rmse}, SNR: {snr}, Gap: {gap_len}s, SNR/RMSE: {quotient}',fontsize=11)
    if i == len(evids_lq)-1:
        ax.set_xlabel('Time (s)',fontsize=11)
    
plt.subplots_adjust(hspace = 0.65,wspace = 0.4)
```
